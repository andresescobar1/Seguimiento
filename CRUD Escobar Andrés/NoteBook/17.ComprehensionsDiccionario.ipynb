## Diccionarios

import os

Este módulo proporciona funciones para interactuar con el sistema operativo, como limpiar la pantalla de la consola. Es útil para mantener la interfaz de usuario ordenada.


cuadrados = {i: i**2 for i in range(1, 101)}

Aquí se utiliza una comprensión de diccionario para crear un diccionario llamado cuadrados, donde:
Las claves son los números del 1 al 100.
Los valores son los cuadrados de esos números (i**2). Esto permite acceder fácilmente al cuadrado de un número.


def mostrar_menu():
    print("Menú:")
    print("1. Ver el cuadrado de un número")
    print("2. Salir")

Esta función imprime el menú de opciones que el usuario puede seleccionar. Facilita la modularidad del código, ya que se puede llamar cada vez que se necesita mostrar el menú.


while True:

Este bucle infinito permite al programa seguir ejecutándose hasta que se elija salir (opción 2). Dentro de este bucle, se ejecutará el menú y se gestionarán las entradas del usuario.


mostrar_menu()
opcion = input("Selecciona una opción (1 o 2): ")

Se llama a mostrar_menu para mostrar las opciones y luego se usa input para capturar la opción seleccionada por el usuario.


if opcion == '1':
    try:
        numero = int(input("Ingresa un número entre 1 y 100: "))
        if 1 <= numero <= 100:
            print(f"El cuadrado de {numero} es {cuadrados[numero]}.")
        else:
            print("Número fuera de rango. Debe estar entre 1 y 100.")
    except ValueError:
        print("Por favor, ingresa un número válido.")

Condición: Si el usuario elige la opción '1':
Intenta convertir la entrada en un número entero.
Verifica si el número está en el rango de 1 a 100.
Si está en rango, imprime el cuadrado usando el diccionario cuadrados.
Si no, informa que el número está fuera de rango.
Si ocurre un ValueError (por ejemplo, si el usuario ingresa texto en lugar de un número), se muestra un mensaje de error.

    os.system('cls' if os.name == 'nt' else 'clear')
    print("Saliendo del programa.")
    break

Si el usuario elige '2', se limpia la pantalla antes de salir y se imprime un mensaje de despedida. El bucle se interrumpe con break.


 else:
    print("Opción no válida. Intenta nuevamente.")

Si el usuario ingresa una opción que no es ni '1' ni '2', se informa que la opción no es válida y se le pide que intente nuevamente.


 os.system('cls' if os.name == 'nt' else 'clear')

Al final de cada iteración del bucle (excepto antes de salir), se limpia la pantalla para que el menú y los mensajes no se acumulen, manteniendo la interfaz limpia.


```python
import os

# Crear el diccionario que asocia cada número con su cuadrado
cuadrados = {i: i**2 for i in range(1, 101)}

def mostrar_menu():
    print("Menú:")
    print("1. Ver el cuadrado de un número")
    print("2. Salir")

while True:
    mostrar_menu()
    opcion = input("Selecciona una opción (1 o 2): ")

    if opcion == '1':
        try:
            numero = int(input("Ingresa un número entre 1 y 100: "))
            if 1 <= numero <= 100:
                print(f"El cuadrado de {numero} es {cuadrados[numero]}.")
            else:
                print("Número fuera de rango. Debe estar entre 1 y 100.")
        except ValueError:
            print("Por favor, ingresa un número válido.")
    
    elif opcion == '2':
        # Limpiar la pantalla antes de salir
        os.system('cls' if os.name == 'nt' else 'clear')
        print("Saliendo del programa.")
        break
    
    else:
        print("Opción no válida. Intenta nuevamente.")

    # Limpiar la pantalla después de cada interacción
    os.system('cls' if os.name == 'nt' else 'clear')
```
